```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_proper_names_of_stars#cite_note-IAU-LSN-1'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tables = soup.find_all('table')
```

```python
table = tables[0]
table
```

```python
rows = table.find_all('tr')
column_items = []
for row in rows:
    # Extract cells from each row
    cells = row.find_all(['th', 'td'])
    # Get the text from the specified column index if it exists
    if len(cells) > 2:
        column_items.append(cells[2].get_text(strip=True))

names = column_items[1:]
```

```python
names
```

```python
def write_list_to_txt(file_path, data_list):
    with open(file_path, 'w') as file:
        for item in data_list:
            file.write(f"{item}\n")
write_list_to_txt('names.txt', [name.lower() for name in names])
```
